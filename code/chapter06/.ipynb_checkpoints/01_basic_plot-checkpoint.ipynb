```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = range(100)
Y = [value**2 for value in X]
plt.plot(X, Y)
plt.show()
```

```python
X_1 = range(100)
Y_1 = [np.cos(value) for value in X_1]

X_2 = range(100)
Y_2 = [np.sin(value) for value in X_2]

plt.plot(X_1, Y_1)
plt.plot(X_1, Y_2)
plt.plot(range(100), range(100))
plt.show()
```

```python
fig = plt.figure()
fig.set_size_inches(10, 5)
ax_1 = fig.add_subplot(1,2,1)
ax_2 = fig.add_subplot(1,2,2)

ax_1.plot(X_1, Y_1, c='b')
ax_2.plot(X_2, Y_2, c='g')
plt.show()
```

### Set color
* http://matplotlib.org/2.0.2/api/colors_api.html

```python
X_1 = range(100)
Y_1 = [value for value in X_1]

X_2 = range(100)
Y_2 = [value + 100 for value in X_2]

plt.plot(X_1, Y_1, color='#000000')
plt.plot(X_2, Y_2, c='c')

plt.show()
```

### Set linestyle
* http://matplotlib.org/examples/lines_bars_and_markers/linestyles.html

```python
plt.plot(X_1, Y_1, c="b", linestyle="dashed")
plt.plot(X_2, Y_2, c="r", linestyle="dotted")

plt.show()
```

### Set title & label

```python
plt.plot(X_1, Y_1, c="b", linestyle="dashed")
plt.plot(X_2, Y_2, c="r", linestyle="dotted")

plt.title('$y = \\frac{ax + b}{test}$')
plt.xlabel('$x_{line}$')
plt.ylabel('y_line')

plt.show()
```

### Set annotate

```python
plt.plot(X_1, Y_1, c="b", linestyle="dashed")
plt.plot(X_2, Y_2, c="r", linestyle="dotted")

plt.text(50, 70, "Line_1")
plt.annotate("Line_2", xy=(50, 150), xytext=(20,175)
             , arrowprops=dict(facecolor='black', shrink=0.05))

plt.title('$y = ax + b$')
plt.xlabel('$x_{line}$')
plt.ylabel('y_line')

plt.show()
```

### Set legend

```python
plt.plot(X_1, Y_1, c="b", linestyle="dashed", label='line_1')
plt.plot(X_2, Y_2, c="r", linestyle="dotted", label='line_2')


plt.title('$y = ax + b$')
plt.legend(shadow=True, fancybox=False, loc="lower right")
plt.xlabel('$x_{line}$')
plt.ylabel('y_line')

plt.show()
```

```python
plt.plot(X_1, Y_1, c="b", linestyle="dashed", label='line_1')
plt.plot(X_2, Y_2, c="r", linestyle="dotted", label='line_2')


plt.title('$y = ax + b$')
plt.legend(shadow=True, fancybox=False, loc="lower right")
plt.xlabel('$x_{line}$')
plt.ylabel('y_line')

plt.grid(True, lw=0.5, ls="--", c="0.3")
plt.xlim(-100, 200)
plt.ylim(-200, 200)
plt.savefig("test.png")

plt.show()
```
