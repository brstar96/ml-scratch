```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = range(100)
Y = [value**2 for value in X]
plt.plot(X, Y)
plt.show()
```

```python
X_1 = range(100)
Y_1 = [np.cos(value) for value in X_1]

X_2 = range(100)
Y_2 = [np.sin(value) for value in X_2]

plt.plot(X_1, Y_1)
plt.plot(X_1, Y_2)
plt.plot(range(100), range(100))
plt.show()
```

```python
fig = plt.figure()
fig.set_size_inches(10, 5)
ax_1 = fig.add_subplot(1,2,1)
ax_2 = fig.add_subplot(1,2,2)

ax_1.plot(X_1, Y_1, c='b')
ax_2.plot(X_2, Y_2, c='g')
plt.show()
```

### Set color
* http://matplotlib.org/2.0.2/api/colors_api.html

```python
X_1 = range(100)
Y_1 = [value for value in X_1]

X_2 = range(100)
Y_2 = [value + 100 for value in X_2]

plt.plot(X_1, Y_1, color='#000000')
plt.plot(X_2, Y_2, c='c')

plt.show()
```

### Set linestyle
* http://matplotlib.org/examples/lines_bars_and_markers/linestyles.html

```python
plt.plot(X_1, Y_1, c="b", linestyle="dashed")
plt.plot(X_2, Y_2, c="r", linestyle="dotted")

plt.show()
```

### Set title & label

```python
plt.plot(X_1, Y_1, c="b", linestyle="dashed")
plt.plot(X_2, Y_2, c="r", linestyle="dotted")

plt.title('$y = \\frac{ax + b}{test}$')
plt.xlabel('$x_{line}$')
plt.ylabel('y_line')

plt.show()
```

### Set annotate

```python
plt.plot(X_1, Y_1, c="b", linestyle="dashed")
plt.plot(X_2, Y_2, c="r", linestyle="dotted")

plt.text(50, 70, "Line_1")
plt.annotate("Line_2", xy=(50, 150), xytext=(20,175)
             , arrowprops=dict(facecolor='black', shrink=0.05))

plt.title('$y = ax + b$')
plt.xlabel('$x_{line}$')
plt.ylabel('y_line')

plt.show()
```

### Set legend

```python
plt.plot(X_1, Y_1, c="b", linestyle="dashed", label='line_1')
plt.plot(X_2, Y_2, c="r", linestyle="dotted", label='line_2')


plt.title('$y = ax + b$')
plt.legend(shadow=True, fancybox=False, loc="lower right")
plt.xlabel('$x_{line}$')
plt.ylabel('y_line')

plt.show()
```

### Save image

```python
plt.plot(X_1, Y_1, c="b", linestyle="dashed", label='line_1')
plt.plot(X_2, Y_2, c="r", linestyle="dotted", label='line_2')


plt.title('$y = ax + b$')
plt.legend(shadow=True, fancybox=False, loc="lower right")
plt.xlabel('$x_{line}$')
plt.ylabel('y_line')

plt.grid(True, lw=0.5, ls="--", c="0.3")
plt.xlim(-100, 200)
plt.ylim(-200, 200)
plt.savefig("test.png")

plt.show()
```

### Scatter

```python
data_1 = np.random.rand(512,2)
data_2 = np.random.rand(512,2)

plt.scatter(data_1[:,0], data_1[:,1], c='b', marker='x', alpha=0.7)
plt.scatter(data_2[:,0], data_2[:,1], c='r', marker='p', alpha=0.7)

plt.show()
```

```python
N = 50

x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi*(15*np.random.rand(N))**2

plt.scatter(x, y, s=area, c=colors, alpha=0.5)

plt.show()
```

### Bar chart

```python
data = [[5.,25.,50.,20.],
       [4.,23.,51.,17],
       [6.,22.,52.,19]]

X=np.arange(4)
plt.bar(X+0.00, data[0], color='b', width=0.25)
plt.bar(X+0.25, data[1], color='g', width=0.25)
plt.bar(X+0.50, data[2], color='r', width=0.25)
plt.xticks(X+0.25, ("A","B","C","D"))

plt.show()
```

```python
data = np.array([[5.,25.,50.,20.],
                 [4.,23.,51.,17],
                 [6.,22.,52.,19]])

color_list = ['b','g','r']
data_label = ["A","B","C"]
X = np.arange(data.shape[1])
for i in range(data.shape[0]):
    plt.bar(X, data[i], bottom=np.sum(data[:i], axis=0),
            color=color_list[i], label = data_label[i])
```

```python
A = [5.,30.,45.,22.]
B = [5.,25.,50.,20.]

X = range(4)

plt.bar(X, A, color='b')
plt.bar(X, B, color='r', bottom=60)

plt.show()
```

```python
women_pop = np.array([5,30,45,22])
men_pop = np.array([5,25,50,20])
X = np.arange(4)

plt.barh(X, women_pop, color='r')
plt.barh(X, -men_pop, color='b')

plt.show()
```
